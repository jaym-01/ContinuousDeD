```python
import shutil
from google.colab import files

def save(name_):
    folder_path = f"/content/runs/{name_}"
    zip_path = f"/content/{name_}_run.zip"

    shutil.make_archive(
        base_name=zip_path.replace(".zip", ""),
        format="zip",
        root_dir=folder_path
    )

    files.download(zip_path)
```
